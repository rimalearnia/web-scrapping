```python
#importer les packages 
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#saisir l'url
url="http://feeds.bbci.co.uk/news/rss.xml"
```

```python
from os import spawnl
#importer le code de la page 
reponse=requests.get(url)
soup= BeautifulSoup(reponse.text,"html.parser")
soup
```

```python

#selectionner un article dans le site bbc news ,clique droit ,inspecter et voir de quel classe il s'agit ici c'est item 
#c'est les item qui contiennet des informations 
items=soup.find_all('item')
items
```

```python
item=items[0]
item
```

```python
item.title.string
```

```python
item.title.text
```

```python
item.description.text
```

```python
news_items=[]
for i in items:
  news_i={}
  news_i['title']=i.title.text
  news_i['description']=i.description.text
  news_i['pubdate']=i.pubdate.text
  news_items.append(news_i)
```

```python
news_items
```

```python
#convertir la liste en base de donnees 
df=pd.DataFrame(news_items,columns=['title','description','pubdate'])
```

```python
df.head()
```

```python
#generer le fichier csv contenant la base de donnee 
df.to_csv('web_scraping.csv',index=False,encoding='utf_8')
```
